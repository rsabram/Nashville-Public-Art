```python
schools = pd.read_csv('/Users/rabram/Desktop/NSS/Nashville-Public-Art/data/Metro_Nashville_Public_Schools_Enrollment_and_Demographics.csv')
```

```python
tvaas = pd.read_csv('/Users/rabram/Desktop/NSS/Nashville-Public-Art/data/data_2018_TVAAS_School_Composite.csv')
```

```python
schools['total_enrollment'] = schools.iloc[:,6:21].sum(axis=1)
```

```python
schools = schools.rename(columns = {'Economically Disadvantaged' : 'FRL'})
```

```python
schools['FRL_pct'] = (schools['FRL'] / schools['total_enrollment']) * 100
```

```python
schools_tvaas = pd.merge(schools,tvaas,left_on = 'School Name',right_on = 'School Name')
```

```python
schools_tvaas = schools_tvaas.drop_duplicates('School Name')
```

```python
schools_tvaas = schools_tvaas.rename(columns={'Black or African American' : 'B'})
```

```python
schools_tvaas['B_pct'] = (schools_tvaas['B'] / schools_tvaas['total_enrollment'])*100
```

```python
schools_tvaas = schools_tvaas.drop(schools_tvaas.iloc[:,6:21],axis=1)
```
